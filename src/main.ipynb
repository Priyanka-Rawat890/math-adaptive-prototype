# Math Adventures ‚Äî Adaptive Learning Prototype

### Objective
An adaptive math learning system that adjusts puzzle difficulty based on the learner's performance.

### Features
- 3 Difficulty Levels: Easy, Medium, Hard  
- Tracks correctness and time  
- Adapts difficulty automatically  
- Displays final performance summary


```python
from puzzle_generator import generate_puzzle
from tracker import PerformanceTracker
from adaptive_engine import adjust_difficulty
```

```python
print("Welcome to Math Adventures!")
name = input("Enter your name: ")
difficulty = input("Choose starting level (Easy/Medium/Hard): ").capitalize()

tracker = PerformanceTracker()
```

```python
for _ in range(5):  # 5 puzzles per session
    q, ans = generate_puzzle(difficulty)
    tracker.start_timer()
    try:
        user = float(input(f"Solve: {q} = "))
    except ValueError:
        print("‚ö†Ô∏è Please enter a valid number!")
        continue

    t = tracker.stop_timer()
    is_correct = abs(user - ans) < 0.01
    tracker.log(is_correct, t, difficulty)

    # üîπ Give feedback
    if is_correct:
        print("‚úÖ Correct!")
    else:
        print(f"‚ùå Wrong! The correct answer was {ans}")

    # üîπ Adjust difficulty after each question
    perf = tracker.summary()
    difficulty = adjust_difficulty(perf, difficulty)
    print(f"Next difficulty: {difficulty}\n")
```

```python
summary = tracker.summary()
print("\nüéØ Session Complete!")
print(f"Student: {name}")
print(f"Accuracy: {summary['accuracy']}%")
print(f"Average Time: {summary['avg_time']} seconds")
print(f"Next Recommended Level: {difficulty}")

correct_count = sum(1 for r in tracker.records if r['correct'])
print(f"Total Correct: {correct_count} out of {len(tracker.records)}")
```

### Session Summary
- **Adaptive Logic:** Rule-based  
- **Criteria:** 
  - If accuracy > 80% and avg time < 10s ‚Üí increase difficulty  
  - If accuracy < 50% ‚Üí decrease difficulty  
  - Else ‚Üí keep same  
- **Metrics Tracked:** Correctness, Response Time, Accuracy  
- **Next Step:** Use real user data to further improve model accuracy.
